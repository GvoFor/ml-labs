Step 5. Replace `CET` filed with new text fields for day, month and year.
Note: day and month should be written in 2-digital format.

```python
data[['Year', 'Month', 'Day']] = data['CET'].str.split('-', expand=True)
data.drop(columns=['CET'], inplace=True)
data
```

Step 6. Determine and print out:
1. A number of days with empty `Events` field;
2. The day (and wind speeds in this day) with minimal mean humidity;
3. The months with mean temparature in range from 0 to 5 degrees;

```python
n = data[' Events'].isna().sum()
print(f'Number of days without any event: {n}')
```

```python
min_mean_humidity = data[' Mean Humidity'].min()
result = data[data[' Mean Humidity'] == min_mean_humidity]
result[['Day', ' Max Wind SpeedKm/h', ' Mean Wind SpeedKm/h', ' Mean Humidity']]
```

```python
mean_temperature = data['Mean TemperatureC']
result = data[(mean_temperature >= 0) & (mean_temperature <= 5)]
result[['Month', 'Mean TemperatureC']]
```

Step 9. Plot a pie chart of wind directions (the sector on the chart should correspond to one of the eight directions - north, south, east, west and all the intermediate)

Note: Here I'll asume that East points to the right, so its sector contains degrees in range (-22.5, 22.5], and then degrees increasing in counter-clockwise (so next sector will be North-East, then North and so on)

```python
DIRECTIONS = ['E', 'NE', 'N', 'NW', 'W', 'SW', 'S', 'SE']

def get_dir(deg):
  i = int(((deg + 360) % 360 - 22.5) / 45)
  return DIRECTIONS[i]

data['WindDirDegrees'].apply(get_dir).value_counts().plot.pie()
```

Step 10. Plot on the same chart:
1. The mean maximum temperature for each month of each year
2. The mean minimum dew point for each month of each year.

```python
mean_values = data.groupby(['Year', 'Month'])[['Max TemperatureC', 'Min DewpointC']].mean()
mean_values.plot.line(subplots=True)
```
